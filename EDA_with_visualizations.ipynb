```python
import plotly.express as px
import pandas as pd

# Load the dataset
data_2017 = pd.read_csv('/content/sample_data/2017.csv')

# Create a choropleth map
fig = px.choropleth(data_2017,
                    locations='Country',
                    locationmode='country names',
                    color='Happiness.Score',
                    hover_name='Country',
                    color_continuous_scale=px.colors.sequential.Plasma,
                    title='World Happiness Score 2017')

fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data_path = "/content/sample_data/2017.csv"
data_2017 = pd.read_csv(data_path)

# Correlation Heatmap
corr_matrix = data_2017.corr(numeric_only=True)
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Happiness Factors')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data_path = "/content/sample_data/2017.csv"
data_2017 = pd.read_csv(data_path)

# Top and Bottom Countries Visualization
sorted_data = data_2017.sort_values(by='Happiness.Score', ascending=False)
top_10 = sorted_data.head(10)
bottom_10 = sorted_data.tail(10)
plt.figure(figsize=(15, 10))
plt.subplot(1, 2, 1)
sns.barplot(data=top_10, x='Happiness.Score', y='Country', palette='Greens_r')
plt.title('Top 10 Happiest Countries')
plt.subplot(1, 2, 2)
sns.barplot(data=bottom_10, x='Happiness.Score', y='Country', palette='Reds_r')
plt.title('Bottom 10 Happiest Countries')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data_path = "/content/sample_data/2017.csv"
data_2017 = pd.read_csv(data_path)

# Factor Contribution Visualization
sorted_data = data_2017.sort_values(by='Happiness.Score', ascending=False)
top_10 = sorted_data.head(10)
factors = ['Economy..GDP.per.Capita.', 'Family', 'Health..Life.Expectancy.',
           'Freedom', 'Generosity', 'Trust..Government.Corruption.']
melted_top_10 = top_10.melt(id_vars='Country', value_vars=factors,
                            var_name='Factor', value_name='Value')
plt.figure(figsize=(15, 10))
sns.barplot(data=melted_top_10, x='Value', y='Country', hue='Factor', palette='Set2', errorbar=None)
plt.title('Factor Contribution to Happiness Score (Top 10 Countries)')
plt.legend(title='Factor', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
